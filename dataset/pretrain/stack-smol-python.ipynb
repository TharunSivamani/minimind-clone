```python
# https://huggingface.co/datasets/bigcode/the-stack-smol/tree/main/data/python

import json

INPUT_PATH = "data.json"
OUTPUT_PATH = "pretrain_code.jsonl"

MAX_CHARS = 512  # strict MiniMind-style filtering


def clean_code(code):
    """Normalize code: remove excessive leading newlines only."""
    return code.lstrip("\n")


def convert():
    count = 0

    with open(INPUT_PATH, "r", encoding="utf-8") as fin, \
         open(OUTPUT_PATH, "w", encoding="utf-8") as fout:

        for line in fin:
            row = json.loads(line)

            code = row.get("content", "")
            if not code.strip():
                continue

            # clean code
            code = clean_code(code)

            # length filtering
            if len(code) >= MAX_CHARS:
                continue

            # final formatted output
            entry = {
                "text": f"<|im_start|>{code}<|im_end|>"
            }

            fout.write(json.dumps(entry, ensure_ascii=False) + "\n")
            count += 1

    print(f"✓ Completed. Total samples written: {count}")
    print(f"Output saved to: {OUTPUT_PATH}")


if __name__ == "__main__":
    convert()
```
